#### Aproximar la soluciòn al problema: 
$\frac{dy}{dt} = y-t^2+1$ 

en el intervalo: $0 \leq t \leq 4$

con la condicion inicial: $y(0) = 0.5$

para los valores: $h_t = 1.0,0.5,0.25,0.125 (N_t = 4,8,16,32)$

Resoleverlo utilizando los mètodos:

Euler.

Runge-Kutta de orden 2.

Runge-Kutta de orden 3.

Runge-Kutta de orden 4.

La soluciòn exacta del sistema es: $y(t) = (t+1)^2-0.5e^t$

```python
def mesh (a , b , Nt ):
    """
    Esta funciòn genera el valor del tamaño de paso para el dominio dado.
    
    Paràmetros:
    ----------------------
    Entrada:
        Int. a
        Valor inicial en el tiempo.
        Int. b
        Valor final en el tiempo.
        Int. Nt
        Nùmero de pasos de tiempo.
    Salida:
        Float. ht
        Tamaño de paso.
    """
    ht = (b - a) / Nt
    return ht

def f(t ,y ):
    """
    Esta función genera los valores de la ecuación diferencial.
    Parámetros:
    -----------------------------------
    Entrada:
        Float. t
        Float. y
    Salida:
        Vector que contiene la evaluación de la ecuación diferencial.
    """
    return y - t **2 + 1

def Exacta (t):
    """
    Esta función genera los valores de la solución exacta.
    Parámetros:
    -----------------------------------
    Entrada:
        Float t.
        Vector de tiempo.
    Salida:
        Float.
        Vector solución de la ecuación.
    """
    return (t +1)**2 - 0.5 * np . exp (t)

def Euler (f , t , w , ht ):
    """
    Esta función edita los vectores para aproximarlos a la solución a partir del método de Euler
    Parámetros:
    -----------------------------------
    Entrada:
        Float. f
        Función que genera los valores de la ecuación diferencial.
        Float. t
        Vector de tiempo
        Float. w
        Vector que contiene la condición inicial y que será editado.
        Float. ht
        Tamaño de paso del tiempo
    Salida:
        No aplica.
    """
    for i , val in enumerate (w [0: -1]):
        w[i +1] = w [i] + ht * f( t[i], w[ i ])
        t[i +1] = t [0] + (i +1) * ht

def RK2 (f , t , w , ht ):
    """
    Esta función edita los vectores para aproximarlos a la solución a partir del método de Runge-Kutta 2.
    Parámetros:
    -----------------------------------
    Entrada:
        Float. f
        Función que genera los valores de la ecuación diferencial.
        Float. t
        Vector de tiempo
        Float. w
        Vector que contiene la condición inicial y que será editado.
        Float. ht
        Tamaño de paso del tiempo
    Salida:
        No aplica.
    """
    for i , val in enumerate (w [0: -1]):
        k1 = ht * f(t[ i], w[i ])
        w[i +1] = w [i] + ht * f(t [i] + ht * 0.5 ,
        w[i ] + k1 * 0.5)
        t[i +1] = a + (i +1) * ht

def RK3 (f , t , w , ht ):
    """
    Esta función edita los vectores para aproximarlos a la solución a partir del método de Runge-Kutta 3.
    Parámetros:
    -----------------------------------
    Entrada:
        Float. f
        Función que genera los valores de la ecuación diferencial.
        Float. t
        Vector de tiempo
        Float. w
        Vector que contiene la condición inicial y que será editado.
        Float. ht
        Tamaño de paso del tiempo
    Salida:
        No aplica.
    """
    for i , val in enumerate (w [0: -1]):
        k1 = ht * f(t[i ], w[i ])
        k2 = ht * f(t[i ] + ht /3,w[i ] + k1 / 3)
        k3 = ht * f(t[i ] + 2 * ht / 3,w[i ] + 2 * k2 / 3)
        w[i +1] = w [i] + ( k1 + 3 * k3 ) / 4
        t[i +1] = a + (i +1) * ht

def RK4 (f , t , w , ht ):
    """
    Esta función edita los vectores para aproximarlos a la solución a partir del método de Runge-Kutta 4.
    Parámetros:
    -----------------------------------
    Entrada:
        Float. f
        Función que genera los valores de la ecuación diferencial.
        Float. t
        Vector de tiempo
        Float. w
        Vector que contiene la condición inicial y que será editado.
        Float. ht
        Tamaño de paso del tiempo
    Salida:
        No aplica.
    """
    for i , val in enumerate (w [0: -1]):
        k1 = ht * f(t[i ], w[i ])
        k2 = ht * f(t[i ] + ht /2 , w[i ] + k1 / 2)
        k3 = ht * f(t[i ] + ht /2 , w[i ] + k2 / 2)
        k4 = ht * f(t[i ] + ht , w[ i] + k3 )
        w[i +1] = w [i] + ( k1 + 2* k2 + 2* k3 + k4 ) / 6
        t[i +1] = a + (i +1) * ht
```

### Cálculo para Nt =4

```python
# Número de pasos de tiempo
Nt = 4 # 4 , 8 , 16 , 32
# Valores de inicio y fin en el tiempo
a = 0
b = 4
# Llamado a la función mesh
ht = mesh (a , b , Nt )
# Condición incial
y0 = 0.5
# Vector de tiempo
t = np.linspace (a , b , Nt +1)
# Vectores auxiliares para las funciones
y_eul = np.zeros(Nt+1);
y_rk2 = np.zeros(Nt+1)
y_rk3 = np.zeros(Nt+1)
y_rk4 = np.zeros(Nt+1)
# Aplicación de condiciones iniciales
y_eul[0] = y0
y_rk2[0] = y0
y_rk3[0] = y0
y_rk4[0] = y0
# Llamado de funciones de los métodos numéricos
Euler(f , t , y_eul , ht )
RK2(f , t , y_rk2 , ht )
RK3(f , t , y_rk3 , ht )
RK4(f , t , y_rk4 , ht )
# Cálculo de la solución exacta
texa = np.linspace(a,b,100)
yexa = Exacta(texa)
yp = Exacta(t)
# Cálculo de los errores
e_eul = np.abs ( yp - y_eul )
e_rk2 = np.abs ( yp - y_rk2 )
e_rk3 = np.abs ( yp - y_rk3 )
e_rk4 = np.abs ( yp - y_rk4 )
n_error_eul = np.linalg.norm ( e_eul , 2)
n_error_rk2 = np.linalg.norm ( e_rk2 , 2)
n_error_rk3 = np.linalg.norm ( e_rk3 , 2)
n_error_rk4 = np.linalg.norm ( e_rk4 , 2)
```

```python
Ecuacion = '$EEul = {:.4f}$'.format(n_error_eul) + ', ERK2 = {:.4f}'.format(n_error_rk2) + \
', ERK3 = {:.4f}'.format(n_error_rk3)+ ', ERK4 = {:.4f}'.format(n_error_rk4)
plt.style.use(['Solarize_Light2'])
fig, (ax1) = plt.subplots(1 ,1)
fig.suptitle('Solución y aproximación con $N_t={}$'.format(Nt), fontsize =14)
ax1.plot(texa , yexa , 'k-', lw=3, label = 'Sol. Exacta')
ax1.plot(t , y_eul , 'b>-.', label = '$Euler$')
ax1.plot(t , y_rk2 , 'y>-.', label = '$RK2$')
ax1.plot(t , y_rk3 , 'gx-.', label = '$RK3$')
ax1.plot(t , y_rk4 , 'rx-.', label = '$RK4$')
ax1.set_title( Ecuacion , fontsize =12 , color = 'blue')
ax1.set_xlabel('$t$')
ax1.set_ylabel('$y(t)$')
ax1.legend(loc = 'upper left', ncol =1 , framealpha =0.75, fancybox =True, fontsize =10)
ax1.grid(color = 'w')
plt.subplots_adjust(hspace=0.35)
#plt.savefig('Logistica_Nt_{}.pdf'.format('4,16,64'))
plt.show()
```

```python
Ecuacion = '$EEul = {:.4f}$'.format(n_error_eul) + ', ERK2 = {:.4f}'.format(n_error_rk2)\
+ ', ERK3 = {:.4f}'.format(n_error_rk3)+ ', ERK4 = {:.4f}'.format(n_error_rk4)
plt.style.use(['Solarize_Light2'])
fig, (ax1) = plt.subplots(1 ,1)
fig.suptitle('Errores con $N_t={}$'.format(Nt), fontsize =14)
ax1.plot( e_eul, 'b>-', label = '$Euler$')
ax1.plot( e_rk2, 'y>-', label = '$RK2$')
ax1.plot( e_rk3, 'gx-', label = '$RK3$')
ax1.plot( e_rk4, 'rx-', label = '$RK4$')
ax1.set_title( Ecuacion , fontsize =12 , color = 'blue')
ax1.set_xlabel('$n$')
ax1.set_ylabel('$Error$')
ax1.legend(loc = 'upper left', ncol =1 , framealpha =0.75, fancybox =True, fontsize =10)
ax1.grid(color = 'w')
ax1.set_yscale('log')
ax1.set_xlim(0.9,4)
plt.subplots_adjust(hspace=0.35)
#plt.savefig('Logistica_Nt_{}.pdf'.format('4,16,64'))
plt.show()
```
